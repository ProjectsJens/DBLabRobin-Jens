```python
import KafkaConstants
import subprocess

import numpy as np
from sklearn.cluster import MiniBatchKMeans

import matplotlib.pyplot as plt
from matplotlib import style
from mpl_toolkits.mplot3d import Axes3D

from pyspark import SparkConf, SparkContext
from pickle import dumps
from datetime import datetime as date

print "start"

# init
style.use("ggplot")
plt.rcParams['figure.figsize'] = (10, 10)
sqlContext = SQLContext(sc)

# create temporary table for queries
sqlContext.sql("CREATE TEMPORARY TABLE tmpTable \
                USING " + KafkaConstants.DATA_FORMAT + " \
                OPTIONS (path '" + KafkaConstants.HDFS_DESTINATION + "')")

# query (and filter) movies
df = sqlContext.sql("SELECT * FROM tmpTable \
                     WHERE budget >= " + str(KafkaConstants.MIN_BUDGET) + " \
                     AND revenue >= " + str(KafkaConstants.MIN_REVENUE) + " \
                     AND vote_average >= " + str(KafkaConstants.MIN_VOTE_AVERAGE) + " \
                     AND vote_count >= " + str(KafkaConstants.MIN_VOTES))
    
print ("movies: " + str(df.count()))

# convert dataframe to np array
array = np.array(df.select("budget", "revenue", "vote_average").collect())
```

```python
# clustering with MiniBatchKMeans
kmeans = MiniBatchKMeans(n_clusters=8)
kmeans.fit(array)
labels = kmeans.labels_
```

```python
# plot cluster data in 2D
colors = ["g.","r.","c.","y.", "b.", "m.", "purple", "lime"]

# we want low budget and low revenue
plt.subplot(1, 3, 1)
for i in range(len(array)):
    plt.plot(array[i][0], array[i][1], colors[labels[i]])
plt.xlabel('budget')
plt.ylabel('revenue')

# we want low budget and high vote_average
plt.subplot(1, 3, 2)
for i in range(len(array)):
    plt.plot(array[i][0], array[i][2], colors[labels[i]])
plt.xlabel('budget')
plt.ylabel('vote_average')

# we want low revenue and high vote_average
plt.subplot(1, 3, 3)
for i in range(len(array)):
    plt.plot(array[i][1], array[i][2], colors[labels[i]])
plt.xlabel('revenue')
plt.ylabel('vote_average')

plt.tight_layout()
plt.show()
```

```python
# plot cluster data in 3D
colors = ["g","r","c","y", "b", "m", "purple", "lime"]

ax = plt.subplot(211, facecolor="1.0", projection='3d')
for i in range(len(array)):
    ax.scatter(array[i][0], array[i][1], array[i][2], c=colors[labels[i]], marker='o')
    
ax.set_xlabel('budget')
ax.set_ylabel('revenue')
ax.set_zlabel('vote_average')

ay = plt.subplot(212, facecolor="1.0", projection='3d')
for i in range(len(array)):
    ay.scatter(array[i][1], array[i][0], array[i][2], c=colors[labels[i]], marker='o')
    
ay.set_xlabel('revenue')
ay.set_ylabel('budget')
ay.set_zlabel('vote_average')

plt.tight_layout()
plt.show()
```

```python
# plot centroids in 3D

centroids = kmeans.cluster_centers_
ax = plt.subplot(111, facecolor="1.0", projection='3d')
for i in range(len(centroids)):
    ax.scatter(centroids[i][0], centroids[i][1], centroids[i][2], c=colors[i], marker='x')
    
ax.set_xlabel('budget')
ax.set_ylabel('revenue')
ax.set_zlabel('vote_average')

plt.tight_layout()
plt.show()
```

```python
# ALWAYS CHANGING WITH NEW CLUSTERING!
# we choose the g cluster => high rating (average limit), low budget and low revenue
# position 0 (start at 0) - colors = ["g","r","c","y", "b", "m", "purple", "lime"]
insiderLabel = 0 # ALWAYS CHANGE POSITION HERE!

# save label in hdfs (for kafka consumer)
subprocess.call(["hdfs", "dfs", "-rm", "-R", KafkaConstants.LABEL_PATH]) # remove old label
sc.parallelize([dumps(str(insiderLabel))]).saveAsPickleFile(KafkaConstants.LABEL_PATH) # save new label in hdfs

# save model in hdfs (for kafka consumer)
subprocess.call(["hdfs", "dfs", "-rm", "-R", KafkaConstants.MODEL_PATH]) # remove old model
sc.parallelize([dumps(kmeans)]).saveAsPickleFile(KafkaConstants.MODEL_PATH) # save new model in hdfs

# get insider movies and convert data for mysql database
rdd = df.rdd.filter(lambda row: kmeans.predict([[row.budget, 
                                                 row.revenue, 
                                                 row.vote_average]])[0] == insiderLabel) \
            .map(lambda row: [row.budget, str(row.genres), row.title,
                              date(year=row.year, month=row.month, day=row.day),
                              row.revenue, row.vote_average, row.vote_count])
df = sqlContext.createDataFrame(rdd, KafkaConstants.DB_SCHEMA) # create dataframe with schema of the database

# export movies to external storage (mysql database)
prop = { "user": KafkaConstants.DB_USER,
         "password": KafkaConstants.DB_PW,
         "driver": "com.mysql.jdbc.Driver" }
df.write.jdbc(url=KafkaConstants.DB_URL, 
              table=KafkaConstants.DB_TABLE, 
              mode="overwrite",
              properties=prop)  

# finish
print ("Insider Movies: " + str(rdd.count()))
```
